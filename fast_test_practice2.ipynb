```python
df=pd.read_csv('ecommerce_dataset.csv',header=None, names=['Category','description'])
```

```python
df.head()
```

```python
df.description[0]
```

```python
df.Category.value_counts()
```

```python
df.Category= '_label_' + df.Category
```

```python
df.head()
```

```python
df['category_description']= df.Category +' ' + df.description
df.head()
```

```python
text_1=df.category_description[0]
text_1
```

```python
import re
```

```python
def preprocess(text):
  text= re.sub(r'[^\w\s]',' ', text)
  text= re.sub(r'[ ]+', ' ', text)
  return text.strip().lower()
```

```python
preprocess(text_1)
```

```python

# Check the data type of 'category_description' column
print(df['category_description'].dtypes)

# Convert the column to string if necessary
if df['category_description'].dtypes != 'object':
    df['category_description'] = df['category_description'].astype('str')

```

```python
# Apply the preprocess function
df['category_description']=df['category_description'].astype('str').map(preprocess)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size= 0.2)
```

```python
test.head()
```

```python
train.to_csv('e_commerce.train', columns=['category_description'], header=None, index=False)
test.to_csv('e_commerce.test', columns=['category_description'], header=None, index=False)
```
